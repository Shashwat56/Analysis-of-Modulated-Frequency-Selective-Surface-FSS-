```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('/content/values - Sinusodial Data.csv')

# Extract the values of Magnitude, Time, and Phase from the data
magnitude = data['Magnitude'].values
time = data['Time (ms)'].values
phase = data['Phase'].values

# Calculate the Reflected Signal using the provided formula
reflected_signal = magnitude * np.sin(2 * np.pi * 1.55 * time * pow(10,9) + phase)

# Set the line width for the graph
line_width = 8  # Specify your desired line width here

# Plotting the graph with increased line width
plt.figure(figsize=(5, 3))
plt.plot(time, reflected_signal, color='blue', linewidth=line_width)
plt.xlabel('Time (ms)')
plt.ylabel('Reflected Signal')
plt.title('Reflected Signal vs Time')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

t = np.linspace(0, 2, 1000)  # Generating 1000 values of t between 0 and 2 ms

# Defining A and B based on the given conditions
A = np.where((t >= 0) & (t <= 0.5), 0.25 + 0.20 * np.sin(2 * np.pi * t),
             np.where((t > 0.5) & (t <= 1), 0.25 - 0.30 * np.sin(2 * np.pi * t),
                      np.where((t > 1) & (t <= 1.5), 0.25 + 0.20 * np.sin(2 * np.pi * t),
                               0.25 - 0.30 * np.sin(2 * np.pi * t))))
B = np.where((t >= 0) & (t <= 0.5), -25 + 75 * np.sin(2 * np.pi * t),
             np.where((t > 0.5) & (t <= 1), -25 + 100 * np.sin(2 * np.pi * t),
                      np.where((t > 1) & (t <= 1.5), -25 + 75 * np.sin(2 * np.pi * t),
                               -25 + 100 * np.sin(2 * np.pi * t))))

Y = A * np.sin(2 * np.pi * t + B)  # Calculating Y

# Plotting Y vs t
plt.subplot(2, 1, 1)
plt.plot(t, Y)
plt.xlabel('t (ms)')
plt.ylabel('Y')
plt.title('Graph of Y vs t')
plt.grid(True)

# Performing Fourier transform
dt = t[1] - t[0]  # Time step
N = len(Y)  # Number of data points
f = fftfreq(N, dt)  # Frequencies
Y_fft = fft(Y)

# Plotting the Fourier transform of Y vs F (GHz)
plt.subplot(2, 1, 2)
plt.plot(f / 1e9, np.abs(Y_fft))
plt.xlabel('F (GHz)')
plt.ylabel('Y_fft')
plt.title('Fourier Transform of Y')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range in milliseconds
t = np.linspace(0, 2, 1000)

# Define A and B based on the given conditions
A = np.where(((t >= 0) & (t <= 0.5)) | ((t > 1) & (t <= 1.5)), 0.25 + 0.20 * np.sin(2 * np.pi * t * 1e9),
             np.where(((t > 0.5) & (t <= 1)) | ((t > 1.5) & (t <= 2)), 0.25 - 0.30 * np.sin(2 * np.pi * t),
                      0.0))
B = np.where(((t >= 0) & (t <= 0.5)) | ((t > 1) & (t <= 1.5)), -25 + 75 * np.sin(2 * np.pi * t),
             np.where(((t > 0.5) & (t <= 1)) | ((t > 1.5) & (t <= 2)), -25 + 100 * np.sin(2 * np.pi * t),
                      0.0))

# Calculate Y = A*(Sin(2*pi*(1e9)*t + B))
Y = A * np.sin(2 * np.pi * (1e9) * t + B)

# Define the frequency range for suppression calculation
f_range = np.linspace(1.54990, 1.55001, 100)  # GHz

# Calculate suppression values for the frequency range
suppression_values = []
for f in f_range:
    fc_plus_fb = f * 1e9  # Convert frequency to Hz
    power_fc = np.mean(Y ** 2)  # Power at the desired frequency fc
    power_fc_plus_fb = np.mean((A * np.sin(2 * np.pi * fc_plus_fb * t + B)) ** 2)  # Power at fc + fb
    suppression_db = 10 * np.log10(power_fc / power_fc_plus_fb)  # Calculate suppression in dB
    suppression_values.append(suppression_db)

# Plot the suppression values
plt.plot(f_range, suppression_values)
plt.xlabel('Frequency (GHz)')
plt.ylabel('Suppression (dB)')
plt.title('Suppression vs Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('/content/values - Saw-tooth Data.csv')

# Extract the values of Magnitude, Time, and Phase from the data
magnitude = data['Magnitude'].values
time = data['Time (ms)'].values
phase = data['Phase'].values

# Calculate the Reflected Signal using the provided formula
reflected_signal = magnitude * np.sin(2 * np.pi * 1.55 * time * pow(10, 9) + phase)

# Set the line width for the time domain plot
line_width = 2  # Specify your desired line width here



# Perform Fourier transformation
sampling_rate = 1 / (time[1] - time[0])  # Calculate the sampling rate
freq = np.fft.fftfreq(len(reflected_signal), 1 / sampling_rate)  # Generate the frequency axis
reflected_signal_fft = np.fft.fft(reflected_signal)  # Perform the Fourier transformation

# Set the line width for the frequency domain plot
line_width_fft = 2  # Specify your desired line width here

# Plot the Fourier transformation of the Reflected Signal vs Frequency
plt.figure(figsize=(10, 6))
plt.plot(freq, np.abs(reflected_signal_fft), color='blue', linewidth=line_width_fft)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transformation of Reflected Signal')
plt.grid(True)
plt.show()
```
